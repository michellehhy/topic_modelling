```python
texts = list(df.loc[:, "text"].values)
```

### Pre-process and vectorise 

```python
import gensim
from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import STOPWORDS
```

```python
import nltk
nltk.download('wordnet')
```

```python
clean_text = []
clean_article = []

for text in texts:
  #remove new line character and punctuations 
  text =  text.rstrip('\n')
  text =  re.sub('[^A-Za-z0-9]+', ' ', text)
  # lower case, de-accents, remove words that are less than 3 characters.  Convert the texts into a list of tokens
  for token in gensim.utils.simple_preprocess(text, deacc=True):
    if token not in gensim.parsing.preprocessing.STOPWORDS and not token.isnumeric():
      clean_text.append(token)
  clean_article.append(clean_text)
```

```python
print(clean_article[0])
```

Adding bigram 

```python
bigram = gensim.models.Phrases(clean_article, min_count=5, threshold=100) 
bigram_2 = gensim.models.phrases.Phraser(bigram)
bigram_words = [bigram_2[text] for text in clean_article]
# for i in range(len(lemmatized_texts)):
#   for token in bigram[lemmatized_texts[i]]:
#     if '_' in token:
#       lemmatized_texts[i].append(token)
```

```python
for i in bigram_words:
  for word in i:
    if '_' in word:
      print(word)
```

```python
lemmatizer = WordNetLemmatizer()
lemmatized_texts = [[lemmatizer.lemmatize(element) for element in text] for text in bigram_words]

print(lemmatized_texts[0])
```

Remove common and rare words

```python
dictionary = Dictionary (lemmatized_texts)
print(dictionary)
```

```python
#dictionary.filter_extremes(no_below=10, no_above=0.2)
```

Build bag of words 

```python
corpus = [dictionary.doc2bow(text) for text in lemmatized_texts]
```

```python
print('Number of unique tokens: %d' % len(dictionary))
print('Number of documents: %d' % len(corpus))
```

## LDA Model 

```python
lda_model = gensim.models.LdaMulticore(corpus, num_topics=3, id2word=dictionary, workers=2, passes=2, iterations = 100)
```

```python
for idx, topic in lda_model.print_topics(-1):
    print('Topic: {} \nWords: {}'.format(idx, topic))
```

```python
from gensim.models import CoherenceModel
# Compute Perplexity
print('\nPerplexity: ', lda_model.log_perplexity(corpus))  # a measure of how good the model is. lower the better.

# Compute Coherence Score
coherence_model_lda = CoherenceModel(model=lda_model, texts=lemmatized_texts, dictionary=dictionary, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('\nCoherence Score: ', coherence_lda)
```
